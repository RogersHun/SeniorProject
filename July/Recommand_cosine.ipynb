> ## 빈도수를 가중치로 활용하는 방법

```python
# 각 맥주별로 리뷰의 빈도수 높은순 7개씩 출력 / 추후에 의미없어 보이는건 제거필요해 보임
for count_tokens in count_df['tokens'][:3]:
    top_5 = Counter(count_tokens).most_common(7)
    print(top_5)
    
```

```python
# 각 맥주별로 리뷰의 빈도수 낮은순 10개 출력
for count_tokens in count_df['tokens'][:3]:
    top_5 = sorted(Counter(count_tokens).most_common(), key=lambda x:x[1])[:10]
    print(top_5)
```

```python
# 빈도수 10개 이하인거 날릴려고 했는데 오래걸려서 일단은 전체 빈도수 상위 50개 단어들만 가져온 다음에, 각 맥주별로 단어의 출현 빈도를 나타내는 DF만들기
# 모든 열의 토큰들을 단일 리스트로 펼치기
import itertools
all_tokens = list(itertools.chain.from_iterable(count_df['tokens']))

# 토큰들의 빈도수 계산
token_counts = Counter(all_tokens)

# 상위 50개 단어 추출
top_50_tokens = token_counts.most_common(50)

for token, count in top_50_tokens:
    print(token, count)
```

```python
top_50_tokens
```

```python
import re

# 작은 따옴표로 둘러싸인 단어들 추출하여 저장
extracted_words = [token[0] for token in top_50_tokens if re.search(r"'[^']*'", token[0])]
extracted_tokens = [(word, frequency) for word, frequency in top_50_tokens if word in extracted_words]
```

일단 이거 가지고 코사인 유사도 계산 및 -> 추천시스템 만들기

```python
# 각 요소들을 가중치로 변환하기 위해 스케일러 적용
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
beers_df = pd.DataFrame(scaler.fit_transform(cosine_df),
                        columns=cosine_df.columns,
                        index = cosine_df.index)
```

```python
beers_df
```

> ### cosine similiarity

beers_df의 가중치를 유사도라 생각하고 활용하여(벡터화 되었다는 형식이라 생각하고) 유사 맥주 추천

```python
from sklearn.metrics.pairwise import cosine_similarity

# 코사인 유사도 계산
cosine_similarities = cosine_similarity(beers_df)

# 추천 함수 정의
def recommend_beers(beer_name, num_recommendations=3):
    # 맥주 이름에 해당하는 인덱스 추출
    beer_index = beers_df.index.get_loc(beer_name)
    
    # 해당 맥주와 다른 맥주 간의 코사인 유사도 가져오기
    beer_similarities = cosine_similarities[beer_index]
    
    # 유사도가 높은 상위 맥주 인덱스 추출 (자기 자신 제외)
    top_indices = beer_similarities.argsort()[::-1][1:num_recommendations+1]
    
    # 추천된 맥주 출력
    print("Recommended Beers:")
    for index in top_indices:
        similarity_percentage = round(beer_similarities[index] * 100, 2)
        print(f"유사한 맥주 이름 : {beers_df.index[index]}, 유사도 : {similarity_percentage}%")
```

```python
recommend_beers('Tsingtao')
```
