```python
import pandas as pd

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.linear_model import LogisticRegression
import pandas as pd

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv('./beer_tok.csv')
```

```python
df_rg = pd.DataFrame({'name': df['beer_name'], 'review': df['tokens_filtered'], 'Aroma':df['Aroma']})
```

```python
df_rg
```

```python
vectorizer = TfidfVectorizer(min_df=0.0, analyzer='char', sublinear_tf=True, ngram_range=(1,3), max_features=5000)
```

```python
X = vectorizer.fit_transform(df['tokens_filtered'])
y = df['Aroma']
```

```python
model = LogisticRegression()
model.fit(X,y)
```

```python
# 새로운 리뷰 데이터를 특성 벡터화하여 예측 수행
new_review = ['This beer is great']
X_new = vectorizer.transform(new_review)
predicted_aroma = model.predict(X_new)

print(f'Predicted aroma: {predicted_aroma}')
```

```python
# 단어별 영향력 계수 확인

# 각 단어와 그에 해당하는 계수 확인
feature_names = vectorizer.get_feature_names_out()
coefficients = model.coef_

coefficients = model.coef_.tolist()  # 배열을 리스트로 변환
word_importance = dict(zip(feature_names, coefficients))
sorted_word_importance = sorted(word_importance.items(), key=lambda x: x[1], reverse=True)

# 단어별 중요도 출력
for word, importance in sorted_word_importance[:10]:
    print(f'Word: {word}, Importance: {importance}')
```

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV

# 리뷰 데이터와 해당하는 라벨(예: Aroma) 데이터를 사용하여 데이터프레임을 생성한다고 가정
df_lr = pd.DataFrame({'review': df['tokens_filtered'], 'Aroma': df['Aroma']})

# 특성 벡터화
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(df_lr['review'])

# 라벨 벡터
y = df_lr['Aroma']

# 로지스틱 회귀 모델 정의
logistic_regression = LogisticRegression()

# 하이퍼파라미터 그리드 준비
param_grid = {'C': [0.1, 1.0, 10.0]}  # C는 규제 파라미터입니다.

# 그리드 서치 수행
grid_search = GridSearchCV(estimator=logistic_regression, param_grid=param_grid, cv=5)
grid_search.fit(X, y)

# 최적의 모델과 파라미터 출력
best_model = grid_search.best_estimator_
best_params = grid_search.best_params_
print("Best Model:", best_model)
print("Best Parameters:", best_params)

# 새로운 리뷰 데이터를 특성 벡터화하여 예측 수행
new_review = ['this beer\'s aroma is so spicy']
X_new = vectorizer.transform(new_review)
predicted_aroma = best_model.predict(X_new)

print(f'Predicted aroma: {predicted_aroma}')
```

```python
# 단어별 영향력 계수 확인

# 각 단어와 그에 해당하는 계수 확인
feature_names = vectorizer.get_feature_names_out()
coefficients = best_model.coef_[0]

word_importance = dict(zip(feature_names, coefficients))
sorted_word_importance = sorted(word_importance.items(), key=lambda x: x[1], reverse=True)

# 단어별 중요도 출력
for word, importance in sorted_word_importance[:10]:
    print(f'Word: {word}, Importance: {importance}')
```

네, 주어진 코드에서 단어별 영향력 계수 확인 부분은 로지스틱 회귀분석을 통해 학습된 모델에서 각 단어의 영향력을 나타내는 계수를 확인하는 부분입니다. 이 계수는 해당 단어가 Aroma 라벨에 대해 얼마나 영향을 미치는지를 나타냅니다.

예를 들어, 양수의 계수는 해당 단어가 Aroma 라벨의 양성(positive) 값을 예측하는 데 영향을 주는 것을 의미합니다. 음수의 계수는 해당 단어가 Aroma 라벨의 음성(negative) 값을 예측하는 데 영향을 주는 것을 의미합니다. 계수의 절댓값이 클수록 해당 단어가 Aroma 라벨 예측에 더 큰 영향력을 가진다고 해석할 수 있습니다.

따라서, sorted_word_importance를 출력하여 단어별 중요도를 확인할 수 있습니다. 결과적으로, 해당 코드는 리뷰 데이터에서 Aroma 라벨에 영향을 주는 단어를 확인하고 그 영향력을 출력하는 역할을 합니다.
